```r
camino<-system.file("data","morley.tab", package = "datasets")
camino
```

```r
a<-file.show(camino)
a
```

```r
mm<-read.table(camino)
mm
```

```r
mm$Expt<-factor(mm$Expt)
mm$Run<-factor(mm$Run)
attach(mm)
```

```r
library(repr)
options(repr.plot.width=5, repr.plot.height=4)
plot(Expt,Speed, main="vel. de la luz", xlab="experimento", col="red", lc="green")
```

```r
plot(Run,Speed, main="vel. de la luz", xlab="Intento")
```

```r
analisis1<-aov(Speed~Run+Expt, data=mm)
```

```r
summary(analisis1)
```

```r
analisis2<-update(analisis1, .~.-Run)
```

```r
tururu<-anova(analisis1,analisis2)
```

summary(analisis2)

```r
plot(analisis1)
```

```r
summary(tururu)
```
